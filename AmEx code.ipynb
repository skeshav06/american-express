```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
#decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid2 = {'max_leaf_nodes': list(range(2, 100)),'min_samples_split': [2, 3, 4]}

#Creating the classifier
model2 = DecisionTreeClassifier(random_state=2)

grid_search2 = GridSearchCV(model2, param_grid=param_grid2, cv=5, scoring='recall', verbose=4)
grid_search2.fit(x_train, y_train)
```

```python
print(grid_search2.best_score_)
print(grid_search2.best_params_)
```

```python
y_pred2 = grid_search2.predict(x_test)
print(accuracy_score(y_test,y_pred2))
```
